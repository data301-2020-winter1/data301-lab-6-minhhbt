```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid',font_scale=2)

df = sns.load_dataset("penguins")
```

```python
vp = sns.violinplot(data=df,x='species',y='bill_length_mm')
plt.figure(figsize=(20,5))
plt.xlabel('Species')
plt.ylabel('Bill Length')
plt.title('Penguin Bill Length by species')
sns.despine()
vp.get_figure().savefig('parsons1.png')
```

```python
plt.figure(figsize=(20,5))
vp1 = sns.violinplot(data=df,x='species',y='bill_length_mm', hue="sex",split="true")
plt.xlabel('Species')
plt.ylabel('Bill Length')
plt.title('Penguin Bill Length by species and sex')
sns.despine()
```
